```python
df['Precio gasolina 95 E5'] = df['Precio gasolina 95 E5'].str.replace(',', '.').astype(float)
df['Precio gasolina 95 E10'] = df['Precio gasolina 95 E10'].str.replace(',', '.').astype(float)
df['Precio gasolina 95 E5 Premium'] = df['Precio gasolina 95 E5 Premium'].str.replace(',', '.').astype(float)
df['Precio gasolina 98 E5'] = df['Precio gasolina 98 E5'].str.replace(',', '.').astype(float)
df['Precio gasolina 98 E10'] = df['Precio gasolina 98 E10'].str.replace(',', '.').astype(float)
df['Precio gasóleo A'] = df['Precio gasóleo A'].str.replace(',', '.').astype(float)
df['Precio gasóleo Premium'] = df['Precio gasóleo Premium'].str.replace(',', '.').astype(float)
df['Precio gasóleo B'] = df['Precio gasóleo B'].str.replace(',', '.').astype(float)
df['Precio gasóleo C'] = df['Precio gasóleo C'].str.replace(',', '.').astype(float)
df['Precio bioetanol'] = df['Precio bioetanol'].str.replace(',', '.').astype(float)
#df['% bioalcohol'] = df['% bioalcohol'].str.replace(',', '.').astype(float)
df['Precio biodiésel'] = df['Precio biodiésel'].str.replace(',', '.').astype(float)
#df['% éster metílico'] = df['P% éster metílico'].str.replace(',', '.').astype(float)
df['Precio gases licuados del petróleo'] = df['Precio gases licuados del petróleo'].str.replace(',', '.').astype(float)
df['Precio gas natural comprimido'] = df['Precio gas natural comprimido'].str.replace(',', '.').astype(float)
df['Precio gas natural licuado'] = df['Precio gas natural licuado'].str.replace(',', '.').astype(float)
#df['Precio hidrógeno'] = df['Precio hidrógeno'].str.replace(',', '.').astype(float)
```

```python
df.dtypes
```

```python
# Vamos a crear una variable nueva, donde tengamos el precio medio de todos los carburantes por gasolinera para
# poder identificar las más baratas y caras:
df['mean_prices'] = df.iloc[:, 9:26].mean(axis=1)
```

```python
# Revisamos las gasolineras más caras
df.sort_values(['mean_prices'], ascending=[False]).head()
```

```python
# Revisamos las gasolineras más baratas
df.sort_values(['mean_prices'], ascending=[False]).tail()
```

```python
# Vemos cuantos carburantes hay por gasolinera. 17 es el total de carburantes disponibles
df['num_carburantes'] = 17 - df.iloc[:, 9:26].isnull().sum(axis=1)
```
